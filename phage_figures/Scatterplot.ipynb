```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## <1> Creates the scatter plot with an overlayed regression line for 1 species
- Date : 23/06/25
- Author : Jeesu Lee

### Acinetobacter baumannii

### <1-1> scatter plot

##### step1. loads the CSV file into a DataFrame

```python
csv_Acinetobacter_baumannii = "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Acinetobacter baumannii.csv"
df_Acinetobacter_baumannii = pd.read_csv(csv_Acinetobacter_baumannii)
```

##### step2. replaces 0 values with NaN

```python
df_Acinetobacter_baumannii.replace(0, float('nan'), inplace=True)
df_Acinetobacter_baumannii.dropna(subset=['Genome lengths', 'Number of ORF'], inplace=True)
```

##### step3. extracts the necessary columns

```python
phage_IDs_Acinetobacter_baumannii = df_Acinetobacter_baumannii['phage ID']
Number_of_ORF_Acinetobacter_baumannii = df_Acinetobacter_baumannii['Number of ORF']
Genome_lengths_Acinetobacter_baumannii = df_Acinetobacter_baumannii['Genome lengths']
```

##### step4. creates a scatter plot

```python
plt.scatter(Genome_lengths_Acinetobacter_baumannii, Number_of_ORF_Acinetobacter_baumannii, c='#1f77b4')
plt.xlabel('Genome lengths')
plt.ylabel('Number of Open reading frames')
    
plt.show()
```

### <1-2> regression line

##### step1. converts the data to NumPy arrays

```python
X = np.array(Genome_lengths_Acinetobacter_baumannii).reshape(-1, 1)
y = np.array(Number_of_ORF_Acinetobacter_baumannii)
```

##### step2. creates and fits the linear regression model

```python
regression_model = LinearRegression()
regression_model.fit(X, y)
```

##### step3. rtrieves the coefficients and intercept of the regression model
##### step4. equation of the regression line

```python
coefficient = regression_model.coef_[0]
intercept = regression_model.intercept_
equation_text = "y = {:.2f}x + {:.2f}".format(coefficient, intercept)
```

##### step5. calculates the predicted values of the regression line

```python
y_pred = regression_model.predict(X)
```

##### step6. calculates the R² value

```python
r2 = r2_score(y, y_pred)
r2_text = "R² = {:.2f}".format(r2)
```

##### step7. creates the scatter plot

```python
plt.figure(figsize=(8, 7))

plt.scatter(Genome_lengths_Acinetobacter_baumannii, Number_of_ORF_Acinetobacter_baumannii, label='Acinetobacter baumannii', c='#1f77b4')
plt.plot(Genome_lengths_Acinetobacter_baumannii, y_pred, color='black')
plt.xlabel('Genome lengths')
plt.ylabel('Number of Open reading frames')

plt.text(0.02, 0.98, equation_text, transform=plt.gca().transAxes, fontsize=10, va='top')
plt.text(0.02, 0.94, r2_text, transform=plt.gca().transAxes, fontsize=10, va='top')
plt.legend(loc='upper right')

plt.tight_layout()
plt.savefig("D:\Bacteriophage\ESKAPE\Regression curve\Scatter_plot_Acinetobacter_baumannii.jpg")

plt.show()
```

### Escherichia coli

```python
csv_Escherichia_coli = "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Escherichia coli.csv"
df_Escherichia_coli = pd.read_csv(csv_Escherichia_coli)

df_Escherichia_coli.replace(0, float('nan'), inplace=True)
df_Escherichia_coli.dropna(subset=['Genome lengths', 'Number of ORF'], inplace=True)

phage_IDs_Escherichia_coli = df_Escherichia_coli['phage ID']
Number_of_ORF_Escherichia_coli = df_Escherichia_coli['Number of ORF']
Genome_lengths_Escherichia_coli = df_Escherichia_coli['Genome lengths']

X = np.array(Genome_lengths_Escherichia_coli).reshape(-1, 1)
y = np.array(Number_of_ORF_Escherichia_coli)

regression_model = LinearRegression()
regression_model.fit(X, y)

coefficient = regression_model.coef_[0]
intercept = regression_model.intercept_
equation_text = "y = {:.2f}x + {:.2f}".format(coefficient, intercept)

y_pred = regression_model.predict(X)

r2 = r2_score(y, y_pred)
r2_text = "R² = {:.2f}".format(r2)

plt.figure(figsize=(8, 7))

plt.scatter(Genome_lengths_Escherichia_coli, Number_of_ORF_Escherichia_coli, label='Escherichia coli', c='#2ca02c')
plt.plot(Genome_lengths_Escherichia_coli, y_pred, color='black')
plt.xlabel('Genome lengths')
plt.ylabel('Number of Open reading frames')

plt.text(0.02, 0.98, equation_text, transform=plt.gca().transAxes, fontsize=10, va='top')
plt.text(0.02, 0.94, r2_text, transform=plt.gca().transAxes, fontsize=10, va='top')
plt.legend(loc='upper right')

plt.tight_layout()
plt.savefig("D:\Bacteriophage\ESKAPE\Regression curve\Scatter_plot_Escherichia_coli.jpg")

plt.show()
```

### <2-2> regression line

##### step1. imports the csv file

```python
csv_files = [
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Acinetobacter baumannii.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Campylobacter jejuni.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Escherichia coli.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Enterococcus faecium.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Haemophilus influenzae.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Helicobacter pylori.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Pseudomonas aeruginosa.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Staphylococcus aureus.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Streptococcus pneumoniae.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Salmonella enterica.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Shigella flexneri.csv"
]

scatter_labels = [
    "Acinetobacter baumannii",
    "Campylobacter jejuni",
    "Escherichia coli",
    "Enterococcus faecium",
    "Haemophilus influenzae",
    "Helicobacter pylori",
    "Pseudomonas aeruginosa",
    "Staphylococcus aureus",
    "Streptococcus pneumoniae",
    "Salmonella enterica",
    "Shigella flexneri"
]

colors = ['#1f77b4',
        '#ff7f0e',
        '#2ca02c',
        '#d62728', 
        '#9467bd',
        '#8c564b', 
        '#e377c2',
        '#7f7f7f',
        '#bcbd22',
        '#17becf',
        '#8c6d31'
]
```

##### step2. initializes the empty list for combined data

```python
combined_genome_lengths = []
combined_number_of_orf = []
```

##### step3 ~ step11

```python
## step3. read and combine data from all CSV files
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df.replace(0, float('nan'), inplace=True)
    df.dropna(subset=['Genome lengths', 'Number of ORF'], inplace=True)
    combined_genome_lengths.extend(df['Genome lengths'])
    combined_number_of_orf.extend(df['Number of ORF'])

## step4. converts the list to NumPy arrays
combined_genome_lengths = np.array(combined_genome_lengths).reshape(-1, 1)
combined_number_of_orf = np.array(combined_number_of_orf)

## step5. creates and fits the linear regression model
regressor = LinearRegression()
regressor.fit(combined_genome_lengths, combined_number_of_orf)

## step6. calculates the predicted values of the regression line
plt.figure(figsize=(8, 7))
x_range = np.linspace(min(combined_genome_lengths), max(combined_genome_lengths), 100)
y_pred = regressor.predict(x_range.reshape(-1, 1))

## step7. creates the scatter plot for all datasets
for i, csv_file in enumerate(csv_files):
    df = pd.read_csv(csv_file)
    df.replace(0, float('nan'), inplace=True)
    df.dropna(subset=['Genome lengths', 'Number of ORF'], inplace=True)
    plt.scatter(df['Genome lengths'], df['Number of ORF'], label=scatter_labels[i],  color=colors[i])

plt.plot(x_range, y_pred, color='black')

## step8. rtrieves the coefficients and intercept of the regression model
## step9. equation of the regression line
equation = f'y = {regressor.coef_[0]:.2f}x + {regressor.intercept_:.2f}'
plt.text(0.02, 0.98, equation, transform=plt.gca().transAxes, fontsize=10, va='top')

## step10. calculates the R² value
r2 = f'R² = {regressor.score(combined_genome_lengths, combined_number_of_orf):.2f}'
plt.text(0.02, 0.94, r2, transform=plt.gca().transAxes, fontsize=10, va='top')

## step11. creats the graph
plt.xlabel('Genome lengths')
plt.ylabel('Number of Open reading frames')
plt.legend()

plt.tight_layout()
plt.savefig("D:\Bacteriophage\ESKAPE\Regression curve\Regression_line_every_host_ESKAPE_pathogen.jpg")

plt.show()
```

```python
csv_files = [
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Acinetobacter baumannii.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Campylobacter jejuni.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Enterococcus faecium.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Haemophilus influenzae.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Helicobacter pylori.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Pseudomonas aeruginosa.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Staphylococcus aureus.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKPAE_Streptococcus pneumoniae.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Salmonella enterica.csv",
    "D:\Bacteriophage\ESKAPE\Regression curve\ESKAPE_Shigella flexneri.csv"
]

# List of custom labels for each scatter plot
scatter_labels = [
    "Acinetobacter baumannii",
    "Campylobacter jejuni",
    "Enterococcus faecium",
    "Haemophilus influenzae",
    "Helicobacter pylori",
    "Pseudomonas aeruginosa",
    "Staphylococcus aureus",
    "Streptococcus pneumoniae",
    "Salmonella enterica",
    "Shigella flexneri"
]

colors = ['#1f77b4',
        '#ff7f0e',
        '#d62728', 
        '#9467bd',
        '#8c564b', 
        '#e377c2',
        '#7f7f7f',
        '#bcbd22',
        '#17becf',
        '#8c6d31'
]

# Initialize empty lists for combined data
combined_genome_lengths = []
combined_number_of_orf = []

# Read and combine data from all CSV files
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df.replace(0, float('nan'), inplace=True)
    df.dropna(subset=['Genome lengths', 'Number of ORF'], inplace=True)
    combined_genome_lengths.extend(df['Genome lengths'])
    combined_number_of_orf.extend(df['Number of ORF'])

# Convert lists to NumPy arrays
combined_genome_lengths = np.array(combined_genome_lengths).reshape(-1, 1)
combined_number_of_orf = np.array(combined_number_of_orf)

# Fit a linear regression model
regressor = LinearRegression()
regressor.fit(combined_genome_lengths, combined_number_of_orf)

# Generate predictions using the regression model
plt.figure(figsize=(8, 7))
x_range = np.linspace(min(combined_genome_lengths), max(combined_genome_lengths), 100)
y_pred = regressor.predict(x_range.reshape(-1, 1))

# Scatter plot for all datasets
for i, csv_file in enumerate(csv_files):
    df = pd.read_csv(csv_file)
    df.replace(0, float('nan'), inplace=True)
    df.dropna(subset=['Genome lengths', 'Number of ORF'], inplace=True)
    plt.scatter(df['Genome lengths'], df['Number of ORF'], label=scatter_labels[i],  color=colors[i])

# Plot the regression line
plt.plot(x_range, y_pred, color='black')

# Equation of the regression line
equation = f'y = {regressor.coef_[0]:.2f}x + {regressor.intercept_:.2f}'
plt.text(0.02, 0.98, equation, transform=plt.gca().transAxes, fontsize=10, va='top')

# R-squared value
r_squared = f'R² = {regressor.score(combined_genome_lengths, combined_number_of_orf):.2f}'
plt.text(0.02, 0.94, r_squared, transform=plt.gca().transAxes, fontsize=10, va='top')

# Configure the graph
plt.xlabel('Genome lengths')
plt.ylabel('Number of Open reading frames')
plt.legend()

# Display the graph
plt.tight_layout()
plt.savefig("D:\Bacteriophage\ESKAPE\Regression curve\Regression_line_10_host_ESKAPE_pathogen.jpg")

plt.show()
```

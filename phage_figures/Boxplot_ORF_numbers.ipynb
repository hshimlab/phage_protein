```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## <1> Creates the boxplot of 11 species : Number of open reading frames
- Date : 23/06/24
- Author : Jeesu Lee

##### step1. creates the boxplot

```python
csv_ORF = "D:\Bacteriophage\ESKAPE\Boxplot\Statistics_all_ORF_final.csv"
df_ORF = pd.read_csv(csv_ORF)
```

```python
df_ORF.replace(0, float('nan'), inplace=True)
```

```python
plt.figure(figsize=(15, 6))

colors = ['#1f77b4',
        '#ff7f0e',
        '#2ca02c',
        '#d62728', 
        '#9467bd',
        '#8c564b', 
        '#e377c2',
        '#7f7f7f',
        '#bcbd22',
        '#17becf',
        '#8c6d31'
]

sns.boxplot(data=df_ORF, palette=colors)

plt.xlabel('Species')
plt.ylabel('Number of open reading frames')
plt.title('Boxplot of open reading frames by the host ESKAPE pathogen')

plt.xticks(rotation=20, ha="right")

plt.tight_layout()
plt.savefig("D:\Bacteriophage\ESKAPE\Boxplot\Figures\Boxplot of open reading frames by the host ESKAPE pathogen.jpg")
```

##### step2. calculates the normalized values

```python
df_ORF_norm = (df_ORF - df_ORF.min()) / (df_ORF.max() - df_ORF.min())
df_ORF_norm
```

```python
plt.figure(figsize=(15, 6))

colors = ['#1f77b4',
        '#ff7f0e',
        '#2ca02c',
        '#d62728', 
        '#9467bd',
        '#8c564b', 
        '#e377c2',
        '#7f7f7f',
        '#bcbd22',
        '#17becf',
        '#8c6d31'
]

sns.boxplot(data=df_ORF_norm, palette=colors)

plt.xlabel('Species')
plt.ylabel('Number of open reading frames')
plt.title('Boxplot of open reading frames by the host ESKAPE pathogen')

plt.xticks(rotation=20, ha="right")

plt.tight_layout()
plt.savefig(r"D:\Bacteriophage\ESKAPE\Boxplot\Figures\Normalize_boxplot of open reading frames by the host ESKAPE pathogen.jpg")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## <1> Creates the boxplot of 11 species : Genome lengths
- Date : 23/06/24
- Author : Jeesu Lee

##### step1. creates the boxplot

```python
csv_genome_lengths = "D:\Bacteriophage\ESKAPE\Boxplot\Statistics_all_Genome_lengths_final.csv"
df_genome_lengths = pd.read_csv(csv_genome_lengths)
```

```python
df_genome_lengths.replace(0, float('nan'), inplace=True)
```

```python
plt.figure(figsize=(15, 6))

colors = ['#1f77b4',
        '#ff7f0e',
        '#2ca02c',
        '#d62728', 
        '#9467bd',
        '#8c564b', 
        '#e377c2',
        '#7f7f7f',
        '#bcbd22',
        '#17becf',
        '#8c6d31'
]

sns.boxplot(data=df_genome_lengths, palette=colors)

plt.xlabel('Species')
plt.ylabel('Phage genome lengths (bp)')
plt.title('Boxplot of phage genome lengths (bp) by the host ESKAPE pathogen')

plt.xticks(rotation=20, ha="right")

plt.tight_layout()
plt.savefig("D:\Bacteriophage\ESKAPE\Boxplot\Figures\Boxplot of phage genome lengths (bp) by the host ESKAPE pathogen.jpg")
```

##### step2. calculates the normalized values

```python
df_genome_lengths_norm = (df_genome_lengths - df_genome_lengths.min()) / (df_genome_lengths.max() - df_genome_lengths.min())
df_genome_lengths_norm
```

```python
plt.figure(figsize=(15, 6))

colors = ['#1f77b4',
        '#ff7f0e',
        '#2ca02c',
        '#d62728', 
        '#9467bd',
        '#8c564b', 
        '#e377c2',
        '#7f7f7f',
        '#bcbd22',
        '#17becf',
        '#8c6d31'
]

sns.boxplot(data=df_genome_lengths_norm, palette=colors)

plt.xlabel('Species')
plt.ylabel('Phage genome lengths (bp)')
plt.title('Boxplot of phage genome lengths (bp) by the host ESKAPE pathogen')

plt.xticks(rotation=20, ha="right")

plt.tight_layout()
plt.savefig(r"D:\Bacteriophage\ESKAPE\Boxplot\Figures\Normalize_boxplot of phage genome lengths (bp) by the host ESKAPE pathogen.jpg")
```
